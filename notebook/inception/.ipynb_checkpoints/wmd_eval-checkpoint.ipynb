```python
model_name="zai-org/GLM-4.6"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote=True)
```

```python
tokenizer.chat_template
```

```python
def get_example_prompt(model_name:str):
    tokenizer=AutoTokenizer.from_pretrained(model_name)
    problem = "Convert the point $(0,3)$ in rectangular coordinates to polar coordinates. Enter your answer in the form $(r,\theta),$ where $r > 0$ and $0 \le \theta < 2 \pi.$"
    tokenized_ = tokenizer.apply_chat_template(
        [{"role":"user", "content": problem}],
        tokenize=False,
        enable_thinking=True,
        add_generation=True
    )
    return tokenized_
```

```python
prompt = get_example_prompt(model_name)
```

```python
print(prompt)
```

```python
response = generate_completions(
    input_prompt=prompt,
    model=model_name,
    client_name='deepinfra',
    max_tokens = 1024
)
```

```python
response = response[0][0]
concatenated_response = prompt + response
```
